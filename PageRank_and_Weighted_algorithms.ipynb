# Класс, высчитывающий pagerank каждой вершины, работающий с ориентированными и не ориентированными графами

# Функция для генерации ориентированных графов с фиксированным количеством связей и неизвестным количеством вершин


```python
def random_digraph(n):
    DG = nx.DiGraph()
    count = n
    i = 0
    while i < count:

        node_a = randint(1, n//2)
 
        node_b = randint(1, n//2)
        #print(node_a, "---", node_b)
        if node_a == node_b:
            i +=1
            continue
        if not(DG.has_edge(node_a, node_b)):
            
            print("добавить", node_a, "---", node_b)
            DG.add_edge(node_a, node_b)
            #if val_a >= val_b:
            DG.add_path([node_a, node_b])
            i +=1


    return DG
```

# Функция для генерации ориентированных графов с фиксированным количеством вершин и неизвестным количеством связей

```python
def random_digraph_nodes(n):
    DG = nx.DiGraph()
    count = randint(1, n*(n-1))
    for j in range(1, n+1):
        DG.add_node(j)
    i = 0
    while i < count:

        node_a = randint(1, n+1)
 
        node_b = randint(1, n+1)
        if node_a == node_b:
            i +=1
            continue
        
        if not(DG.has_edge(node_a, node_b)):
            
            #print(node_a, "---", node_b)
            DG.add_edge(node_a, node_b)
            #if val_a >= val_b:
            DG.add_path([node_a, node_b])
            i +=1

        


    return DG
```

```python
class PageRank_weight_d_acc:
    def __init__(self, graph, directed, d, acc):
        self.graph = graph
        self.V = len(self.graph)
        self.d = float(d)
        self.directed = directed
        self.ranks = dict()
        self.iterr = 0
        self.acc = acc
        
    def win(self, n):
        
        I_u = len(self.graph.in_edges(n[1]))
        neighb_1 = self.graph.out_edges(n[0])
        P_I = 0
        for nei in neighb_1:
            P_I+= len(self.graph.in_edges(nei[1]))
        
        if (P_I==0) or (I_u==0):
            return 1
        else:
            return float(I_u/P_I)
 
    def wout(self, n):
        O_u = len(self.graph.out_edges(n[1]))
        neighb_1 = self.graph.out_edges(n[0])
        P_O = 0
        for nei in neighb_1:
            P_O+= len(self.graph.out_edges(nei[1]))
        
        if (P_O==0) or (O_u==0):
            return 1
        else:
            return float(O_u/P_O)
        
    
    
    def rank_weight_d_acc(self):
        for key, node in self.graph.nodes(data=True):
            if self.directed:
                self.ranks[key] = 1
            else:
                self.ranks[key] = node.get('rank')

                
        proverka_1 = []
        proverka_2 = []  
        while True:
            self.iterr +=1
            proverka = self.ranks.copy()
            for tup in self.ranks.values(): proverka_1.append(tup)
            for key, node in self.graph.nodes(data=True):
                rank_sum = 0
                curr_rank = node.get('rank')

                neighbors = self.graph.in_edges(key)
                for n in neighbors:
                    rank_sum += self.ranks[n[0]]*self.win(n)*self.wout(n)

            
                self.ranks[key] = ((1 - float(self.d)) * (1/float(self.V))) + self.d*rank_sum
            for tup in self.ranks.values(): proverka_2.append(tup)
            y = 0
            rrr = []
            rrr = list(map(sub, proverka_2, proverka_1))

            for fr in rrr:

                if math.fabs(fr)< self.acc:
                    y+=1

            if y == len(rrr):

                break    
                
            proverka_1 = []
            proverka_2 = []    
        return p
```

```python
def frange(start, stop, step):
    i = start
    while i < stop:
        yield i
        i += step
```

```python
time_all_nodes_weight_fix = []
iterr_all_nodes_weight_fix = []


time_all_nodes_fix = []
iterr_all_nodes_fix = []

var_time_all_nodes_fix = []
var_time_all_nodes_weight_fix = []
var_iterr_all_nodes_fix = []
var_iterr_all_nodes_weight_fix = []

col_endges_fix = []

for i in range(10, 9900, 10):
    time_t_weight = []
    it_t_weight = []
    
    time_t = []
    it_t = []
    print(f"Граф с количеством вершин 100 и количеством связей - {i}")
    for j in range(20):
        
        isDirected = True
        Graph = random_digraph_nodes_fix(i)
        
        
        start = time.perf_counter()
        
        p = PageRank(Graph, isDirected)
        p.rank()
        
        end = time.perf_counter()
        
        time_es = end - start
        iterr_t = p.iterr
        
        it_t.append(iterr_t)
        time_t.append(time_es)
        
        start = time.perf_counter()
        
        p = PageRank_weight(Graph, isDirected)
        p.rank_weight()
        
        
        end = time.perf_counter()
        time_es = end - start
        iterr_t = p.iterr
        
        it_t_weight.append(iterr_t)
        time_t_weight.append(time_es)
    
    time_mean = mean(time_t)
    time_mean_weight = mean(time_t_weight)
    iterr_mean = mean(it_t)
    iterr_mean_weight = mean(it_t_weight)
    
    var_time_all = np.var(time_t)
    var_time_all_weight = np.var(time_t_weight)
    var_iterr_all = np.var(it_t)
    var_iterr_all_weight = np.var(it_t_weight)
    
    var_time_all_nodes_fix.append(var_time_all)
    var_time_all_nodes_weight_fix.append(var_time_all_weight)
    var_iterr_all_nodes_fix.append(var_iterr_all)
    var_iterr_all_nodes_weight_fix.append(var_iterr_all_weight)
    
    print(f"Среднее время выполнения оригинального алгоритма - {time_mean}")
    print(f"Среднее количество итераций оригинального алгоритма - {iterr_mean}")
    print(f"Среднее время выполнения взвешенного алгоритма - {time_mean_weight}")
    print(f"Среднее количество итераций взвешенного алгоритма - {iterr_mean_weight}")
    
    time_all_nodes_weight_fix.append(time_mean_weight)
    iterr_all_nodes_weight_fix.append(iterr_mean_weight)


    time_all_nodes_fix.append(time_mean)
    iterr_all_nodes_fix.append(iterr_mean)
    
    col_endges_fix.append(i)
```
